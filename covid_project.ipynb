## Import Python Libraries

```python
import pandas as pd

from datetime import *
import matplotlib.pyplot as plt
```

Reads the file with all the information on it and converts it into a dataframe. I modified it to always give me yesterday's data

```python
yesterday = datetime.now() - timedelta(days=1)
formatted = yesterday.strftime('%m''-''%d''-''%Y')

covid_data = pd.read_csv(f'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{formatted}.csv')
```

### Displays the top five rows of the dataframe

```python
five = covid_data.head()

five
```

### View Stats by Country

```python
option = input("Pick which country's data you want: ")
choice = covid_data[covid_data["Country_Region"]==option]
while len(choice) < 1:
    print('Invalid option')
    option = input("Pick which country's data you want: ")
    choice = covid_data[covid_data["Country_Region"]==option]

choice
```

### View Region / Statewise Stats for Selected Country

```python
dfplot_region = choice[choice['Deaths'] > 1000]
dfplot_region.sort_values(by='Deaths', ascending = True, inplace = True)
plt.figure(figsize = (10,10), facecolor = 'lightgreen')
ax = plt.subplot()
ax.grid()

ax.barh(dfplot_region['Province_State'], dfplot_region['Deaths'])
```

### Countrywise Stats

```python
countries = covid_data.groupby(by='Country_Region').sum()
    
countries.reset_index(inplace=True)
countries
```

#### Countrywise Deaths exceeding 50,000

```python
dfplot = countries[countries['Deaths'] > 50000]
dfplot.sort_values(by='Deaths', ascending = True, inplace = True)

plt.figure(figsize = (10,10), facecolor = 'Green')
ax = plt.subplot()
ax.grid()
ax.barh(dfplot['Country_Region'],dfplot['Deaths'])
```

### Worldwide Cases over time - Trendline (Last 30 Days)

```python
trend = {"Date":[],'Confirmed':[]}
start = yesterday - timedelta(days=30)
while start <= yesterday:
    
    covid_data = pd.read_csv(f"https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{start.strftime('%m-%d-%Y')}.csv")
    cases = covid_data.Confirmed.sum()
    trend['Confirmed'].append(cases)
    trend['Date'].append(start)
    start = start +  timedelta(days=1)
    
```

```python
df = pd.DataFrame(trend)
```

```python
df["Date"] = df["Date"].dt.strftime('%d-%b-%y')
```

```python
df['Confirmed'] = df['Confirmed'].apply(lambda x : x/1000000)
```

```python
import matplotlib.ticker as tkr
plt.figure(figsize = (16,10))
ax = plt.subplot()
ax.plot(df['Date'], df['Confirmed'],color="red")
interval = tkr.MultipleLocator(3)
ax.xaxis.set_major_locator(interval)
plt.savefig("test.png")
```
