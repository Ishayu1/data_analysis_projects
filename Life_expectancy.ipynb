#### Percentage change in life expectancy from 1960 to 2015


#### Contains data from 1960

```python
df2 = df[(df['Year']== 1960)]
df2
```

#### Contains data from 2015

```python
df3 = df[(df['Year']== 2015)]
df3
```

#### Merged the two dataframes with the first and last data

```python
merged = pd.merge(df2,df3,how="inner",on="Country")
merged
```

```python
merged.drop(['Country Code_y', 'Region_y','Income Group_y'], axis = 1)
```

```python
merged.rename(columns = {'Country Code_x':'Country Code',
                         'Region_x':'Region', 
                         'Income Group_x':'Income Group',
                         'Year_x': 'First Year',
                         'Life Expectancy_x':'Starting life expectancy',
                        'Year_y':'Final Year',
                        'Life Expectancy_y':'Final life expectancy'}, inplace = True)
merged
```

```python
new = merged.drop(['Country Code_y', 'Region_y','Income Group_y'], axis = 1)
```

```python
new['% change'] = new.apply(lambda x: ((x['Final life expectancy'] - x['Starting life expectancy'])/x['Starting life expectancy']) * 100,axis = 1)
new
```

```python
new['% change'].min()
```

```python
new['% change'].max()
```

```python
new.sort_values(by='% change', ascending = True, inplace = True)
top_10 = new[new['% change'] >= 78]
```

```python
years = []
for i in range(0,150,5):
    years.append(i)
plt.figure(figsize = (10,10), facecolor = 'lightgreen')
ax = plt.subplot()
ax.grid()
ax.barh(top_10['Country'], top_10['% change'])

ax.set_title(f" Countries with the highest percentage increase in life expectancy from 1960 - 2015")
ax.set_xlabel('% increase')
ax.set_ylabel('Country')
ax.set_xticks(years)
```

#### World average life expectancy over time

```python
world_av = []
for i in range(1960,2016):
    
    n = df[df['Year'] == i]
    world_av.append(n['Life Expectancy'].mean())
```

```python
world_average = df.groupby(by='Year').mean()
world_average.reset_index(inplace = True)
```

```python
years = []
for i in range(1960,2020,5):
    years.append(i)
plt.figure(figsize = (16,10))

plt.figure(figsize = (10,10), facecolor = 'lightgreen')
ax = plt.subplot()
ax.grid()
ax.plot(world_average['Year'],world_average['Life Expectancy'] ,color="blue",linewidth=2,linestyle='-', 
        markersize=5,marker='o',mfc="black")
ax.set_title(f" Average worldwide life expectancy from 1960 - 2015")
ax.set_xlabel('Year')
ax.set_xticks(years)
ax.set_ylabel("Life expectancy")
ax.legend(loc='upper left', frameon=False)


```
